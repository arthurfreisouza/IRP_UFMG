```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotting KDE with different bandwidths
plt.figure(figsize=(12, 6))

# KDE with default bandwidth
plt.subplot(1, 2, 1)
sns.kdeplot(data, fill=True, color='blue', alpha=0.5)
plt.title('KDE with Default Bandwidth')

# KDE with smaller bandwidth
plt.subplot(1, 2, 2)
sns.kdeplot(data, fill=True, color='red', alpha=0.5, bw_adjust=0.3)
plt.title('KDE with Smaller Bandwidth')

plt.show()
```
