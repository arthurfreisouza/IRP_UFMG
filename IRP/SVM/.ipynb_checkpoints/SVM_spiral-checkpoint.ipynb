```python
import pandas as pd
import numpy as np
import spiral
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 1000, noise = 0.1)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = "viridis")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
matrix = np.concatenate((X, y.reshape(-1, 1)), axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
print(f"Train shape : {X_train.shape}, Test shape : {X_test.shape}")
```

```python
from sklearn.svm import SVC

svm = SVC(C = 0.15)
```

### Fitting the data : 

```python
svm.fit(X_train, y_train)

svm_pred = svm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, svm_pred)
display = ConfusionMatrixDisplay(confusion_matrix = cm)
display.plot()
plt.show()
```

```python
def generate_surface(best_param):  # Define the mesh grid for the two features
    svm = SVC(C = best_param)
    svm.fit(X_train, y_train)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    # Predict the label for each point in the mesh grid
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot the contour and training points
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Surface')
    plt.show()
```

### Making a GS : 

```python
from sklearn.model_selection import GridSearchCV
a = np.arange(start = 0.08, stop = 5, step = 0.25)
kern_type = ["linear", "poly", "rbf", "sigmoid"]
parameters = {"C" : a, "kernel" : kern_type}
GS = GridSearchCV(svm, param_grid = parameters, cv = 10)
GS.fit(X_train, y_train)
print(f"The best accuracy was {GS.best_score_} with the parameters {GS.best_params_}")
```

```python
df_result = pd.DataFrame(data = GS.cv_results_)
df_result.head()
```

```python
for key, value in GS.best_params_.items():
    best_param = value
generate_surface(best_param)
```

```python
def generate_surface(best_param):
    # Treinando o classificador SVM com o melhor parmetro C
    svm = SVC(C=best_param, kernel='linear')  # Adicionando kernel linear
    svm.fit(X_train, y_train)
    
    # Definindo os limites do grfico e criando a malha de pontos
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    
    # Predizendo a classe para cada ponto na malha
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plotando o contorno da superfcie de deciso
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    
    # Plotando os pontos de treinamento
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k')
    
    # Adicionando os vetores de suporte ao grfico
    plt.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], 
                s=100, facecolors='none', edgecolors='k', label='Vetores de Suporte')

    # Plotando as margens e a funo de deciso
    plt.contour(xx, yy, svm.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape), 
                colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'], alpha=0.5)
    
    # Ttulos e rtulos
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Superfcie de Deciso e Vetores de Suporte')
    plt.legend()  # Exibir legenda
    plt.show()
```
