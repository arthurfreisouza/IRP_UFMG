### Testing if data follows a normal pattern.

```python
from scipy import stats
# Using Shapiro-Wilk Test
Shapiro = []

for i in range(data.shape[1]):  # Loop over columns instead of rows
    col_data = data.iloc[:, i].dropna()  # Drop any NaN values
    p_val = stats.shapiro(col_data)[1]
    if p_val >= 0.5:
        print(f"The column {data.iloc[:, i].name} is normally distributed.")
        Shapiro.append(col_data)
if not Shapiro:
    print("There are not Normal Distributions in Dataset.")
else:
    print("There are columns Normally Distributed")
```

```python
# Using Shapiro-Wilk Test

Kolmogorov = []
for i in range(data.shape[1]):  # Loop over columns instead of rows
    col_data = data.iloc[:, i].dropna()  # Drop any NaN values
    p_val = stats.kstest(col_data, "norm")[1]
    if p_val >= 0.5:
        print(f"The column {data.iloc[:, i].name} is normally distributed.")
        Kolmogorov.append(col_data)
if not Kolmogorov:
    print("There are not Normal Distributions in Dataset.")
else:
    print("There are columns Normally Distributed")
```

```python
def plot_graphics(labels: np.ndarray, matrix: np.ndarray):
    N = matrix.shape[1]  # Number of columns in the matrix
    
    # Create subplots with dynamic figsize for N subplots
    fig, axs = plt.subplots(1, N, figsize=(N*5, 5)) 
    
    # Check if N is 1, if so, axs will not be a list, just a single object
    if N == 1:
        axs = [axs]  # Convert to a list for consistent indexing
    
    # Plot each column in the matrix
    for i in range(N):
        axs[i].scatter(labels, matrix[:, i])
        axs[i].set_title(f'Plot {i+1}')
        axs[i].set_xlabel('Labels')
        axs[i].set_ylabel(f'Matrix Column {i+1}')
    
    plt.tight_layout()  # Prevent overlap
    plt.show()
plot_graphics(np.array(labels), np.array(data_scaled))
```

```python
def knn(matrix: np.ndarray, new_points: np.ndarray, K: int = None) -> dict:
    if K is None or K <= 0 or K/2 == 0:
        raise ValueError("The K value must be a positive integer and odd.")
    
    dic = {}
    count = 0

    for point in new_points:
        count += 1
        # Calculate the Euclidean distance between the point and all points in the matrix
        distances = np.linalg.norm(matrix[:, : -1] - point, axis=1)
        
        # Find the indices of the K nearest neighbors
        nearest_indices = np.argsort(distances)[:K]
        
        # Sum the labels (column 2) of the nearest neighbors
        sum_labels = np.sum(matrix[nearest_indices, -1])
        
        # Assign label based on the sum of the nearest neighbors' labels
        dic[f"Points {count}"] = 1 if sum_labels > 0 else -1

    return dic
```

```python
matrix = np.concatenate((X_train, y_train.reshape(-1, 1)), axis = 1)

# Example usage
dic_ret = knn(matrix = matrix, new_points = X_test, K=3)
```

```python
y_pred = np.array([values for key, values in dic_ret.items()]).reshape(-1, 1)
```

# Realizando o CV manualmente, tomando como base o dataset randomizado.

```python
#y = np.array([-1 if i == 0 else 1 for i in labels])
#unique, counts = np.unique(y, return_counts = True)
#print(dict(zip(unique, counts)))
```

```python
def CV():
    X = np.array(data_scaled)
    lst_K = []
    acc = []
    sdv = []
    for K in range(1, 50 + 1):
        X = np.array(data_scaled)
        y = y = np.array([-1 if i == 0 else 1 for i in labels])
        sum_acc = []
        lst_K.append(K)
        for i in range(10):
            np.random.seed(i)
            indices = np.random.permutation(len(labels))
            X = X[indices]
            y = y[indices]
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)
            matrix = np.concatenate((X_train, y_train.reshape(-1, 1)), axis = 1)
            dic_ret = knn(matrix = matrix, new_points = X_test, K = K)
            y_pred = np.array([values for key, values in dic_ret.items()])
            sum_acc.append(accuracy_score(y_test, y_pred))
        acc.append(np.array(sum_acc).mean())
        sdv.append(np.array(sum_acc).std())
    return np.array(acc), np.array(sdv), np.array(lst_K)
```

```python
acc, sdv, lst_K = CV()
```

```python
acc = np.array([round(num, 3) for num in acc])
sdv = np.array([round(num, 3) for num in sdv])
index = [f"K = {i}" for i in range(1, sdv.shape[0] + 1)]
```

```python
plot_results = pd.DataFrame(data = np.concatenate((acc.reshape(-1, 1), sdv.reshape(-1, 1)), axis = 1), columns = ["Mean accuracy", "Standard Deviation"], index = index)
plot_results
```

```python
plt.plot(lst_K, acc, color = "black")
plt.title("Accuracy in function of K")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.show()
```
