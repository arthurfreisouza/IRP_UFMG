```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_sample(size : int, sdv : float, displacement : np.ndarray = None):
    return (np.random.normal(scale = sdv, size = (size,)) + displacement)
```

```python
# Creating the sample 1.
x1l = create_sample(size = 60, sdv = 0.6, displacement = (2, )).reshape(-1, 1)
x1c = create_sample(size = 60, sdv = 0.6, displacement = (2, )).reshape(-1, 1)
x1 = np.concatenate((x1l, x1c), axis = 1)
```

```python
# Creating the sample 2.
x2l = create_sample(size = 60, sdv = 0.4, displacement = (4, )).reshape(-1, 1)
x2c = create_sample(size = 60, sdv = 0.4, displacement = (4, )).reshape(-1, 1)
x2 = np.concatenate((x2l, x2c), axis = 1)
```

```python
def plot_graphic(x1, x2, xrange : np.ndarray = None, f11 = None, f12 = None, f21 = None, f22 = None, gauss = False):
    arr_ = ["red", "blue"]
    plt.scatter(x1[:, 0], x1[:, 1], c=arr_[0])
    plt.scatter(x2[:, 0], x2[:, 1], c=arr_[1])

    if gauss == True:
        # Plot each Gaussian curve if the corresponding parameter is provided
        if f11 is not None:
            plt.plot(xrange, f11, c=arr_[0])
        if f12 is not None:
            plt.plot(f12, xrange, c=arr_[0])
        if f21 is not None:
            plt.plot(xrange, f21, c=arr_[1])
        if f22 is not None:
            plt.plot(f22, xrange, c=arr_[1])
        
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()
```

```python
plot_graphic(x1, x2)
```

```python
def normal_gauss(x, m, r):
    return np.exp(-0.5 * ((x - m) / r) ** 2) / (np.sqrt(2 * np.pi) * r)
```

```python
# Calculating the mean and standard deviation of each sample :

m11 = np.mean(x1l)
m12 = np.mean(x1c)
m21 = np.mean(x2l)
m22 = np.mean(x2c)

s11 = np.std(x1l)
s12 = np.std(x1c)
s21 = np.std(x2l)
s22 = np.std(x2c)
```

```python
xrange = np.linspace(start = 0, stop = 6, num = 60)

f11 = normal_gauss(xrange, m11, s11).reshape(-1, 1)
f12 = normal_gauss(xrange, m12, s12).reshape(-1, 1)
f21 = normal_gauss(xrange, m21, s21).reshape(-1, 1)
f22 = normal_gauss(xrange, m22, s22).reshape(-1, 1)
```

```python
plot_graphic(x1, x2, xrange, f11, f12, f21, f22, gauss = True)
```

```python
lseq = xrange.shape[0]
M1 = np.zeros((lseq, lseq))
M2 = np.zeros((lseq, lseq))
sup = np.zeros((lseq, lseq))
```

```python
for i in range(xrange.shape[0]):
    for j in range(xrange.shape[0]):
        M1[i, j] = f11[i].item() * f12[j].item() # likelihood sample 1.
        M2[i, j] = f21[i].item() * f22[j].item() # likelihood sample 2.
        sup[i, j] = np.sign( M1[i, j] -  M2[i, j])
```

```python
# Ploting the separation surface : 

plt.contour(xrange, xrange, sup, levels = [-1, 1], colors = 'black', linewidths = 4.0)
plt.scatter(x1[:, 0], x1[:, 1], c = "red")
plt.scatter(x2[:, 0], x2[:, 1], c = "blue")
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.show()
```

```python
X, Y = np.meshgrid(xrange, xrange)
```

```python
# Ploting the 3D surface : 
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, M1, cmap = "viridis")

# Setting axis limits and labels
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("(x1, x2)")

# Adjusting view angle to match theta and phi
ax.view_init(elev = 40, azim = 60)  # 'elev' is the equivalent of 'phi', 'azim' of 'theta'

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, M2, cmap = "viridis")

# Setting axis limits and labels
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("(x1, x2)")

# Adjusting view angle to match theta and phi
ax.view_init(elev = 40, azim = 60)  # 'elev' is the equivalent of 'phi', 'azim' of 'theta'

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, M1, cmap = "viridis")
ax.plot_surface(X, Y, M2, cmap = "viridis")


# Setting axis limits and labels
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("(x1, x2)")

# Adjusting view angle to match theta and phi
ax.view_init(elev = 40, azim = 60)  # 'elev' is the equivalent of 'phi', 'azim' of 'theta'

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, sup, cmap = "viridis")

# Setting axis limits and labels
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("(x1, x2)")

# Adjusting view angle to match theta and phi
ax.view_init(elev = 40, azim = 60)  # 'elev' is the equivalent of 'phi', 'azim' of 'theta'

plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, M1, cmap = "viridis")
ax.plot_surface(X, Y, M2, cmap = "viridis")
ax.plot_surface(X, Y, sup, cmap = "viridis")

# Setting axis limits and labels
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("(x1, x2)")

# Adjusting view angle to match theta and phi
ax.view_init(elev = 40, azim = 60)  # 'elev' is the equivalent of 'phi', 'azim' of 'theta'

plt.show()
```
