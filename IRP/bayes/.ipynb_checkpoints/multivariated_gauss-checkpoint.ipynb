```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdfnvar(x : np.ndarray, m : np.ndarray, K : np.ndarray, n):
    diff = x - m
    det_K = np.linalg.det(K)
    inv_K = np.linalg.inv(K)
    exponent = -.5*(diff.T @ inv_K @ diff)
    denominator = np.sqrt((2 * np.pi) ** n * det_K)

    return (1/denominator)*np.exp(exponent)
```

```python
# Parameters
s1 = 1
s2 = 1
ro = 0.8
mean = np.array([3, 3])  # Mean of the distribution
cov = np.array([[s1**2, ro*s1*s2], [ro*s1*s2, s2**2]])  # Covariance matrix

seq = np.linspace(start = 0, stop = 20, num = 200)

seqi, seqj = np.meshgrid(seq, seq)

mat = np.zeros(seqi.shape)
```

```python
for i in range(seqi.shape[0]):
    for j in range(seqi.shape[1]):
        point = np.array([seqi[i, j], seqj[i, j]])
        mat[i, j] = pdfnvar(x = point, m = mean, K = cov, n = 2)
```

```python
plt.contour(seqi, seqj, mat)
plt.xlabel('seqi')
plt.ylabel('seqj')
plt.title('Contour Plot')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.show()
```
